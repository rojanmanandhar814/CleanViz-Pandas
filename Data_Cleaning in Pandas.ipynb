## Data Cleaning in Pandas |

```python
import pandas as pd
```

```python
df = pd.read_excel(r"D:\BootCamp\Python\Web scraping project pandas\Customer Call List.xlsx")
```

```python
df
```

```python
df = df.drop_duplicates()
df
```

```python
#deleting a column as it is not necessary 
df = df.drop(columns = "Not_Useful_Column")
df
```

#### Cleaning & Standaerizing column: Last_Name

```python
# As we are checking each column, the First_Name column seems fine hence we will be standardizing the last_Name column
# specifying the column name is essential 
df["Last_Name"].str.strip()   # Strip is used to remove leading and trailing whitespace (including spaces, tabs, and newlines)
#lstrip meaning left strip removes the spaces on the left side & rstrip removes on the right side 
```

```python
#Method 1 :
df["Last_Name"] = df["Last_Name"].str.lstrip("...") # inside the brackets we need to insert values as to what we want to remove 
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")

df

#Method 2 (All at once)
df["Last_Name"] = df["Last_Name"].str.strip(".../_")
df
```

#### Cleaning & Standaerizing column: Phone_Number

```python
df
#we want all the phone number to look the same 
# we want just number where we will remove the non-numeric values within the columns 
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)  #^ an anchor that matches the start of the string 
df
#regex recognizes the patterns which is set to true 
```

```python
#we will use the lambda function to insert - after every 3 digits 
#lamda defines an anonymous function that inputs parameters and is used for single-line operations that get returned 
# We have to define the index number after which number we want the - 

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# The above error says we have a float and string since in the coloum we have NaN which is a string hence we will convert into all strings first 
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
# As you can see we have the - after every number now 
```

```python
# Now we will remove the Nan-- or nan-- value from the column
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df

df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
df
```

#### Cleaning & Standaerizing column: Address

```python
df[["Street Address","State","Zip_Code"]] = df['Address'].str.split(',', n=2, expand=True)
df
```

#### Cleaning & Standaerizing column: Paying Customer 

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df
```

#### Cleaning & Standaerizing column: Do Not Contact

```python

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

#### Cleaning & Standaerizing column: Removing the N/a or any null values from the entire data sheet

```python
#df = df.replace('N/a','')
#df = df.replace('NaN','')

df.fillna('')
```

#### Listing out only the phone numbers we can call looking at Do not Contact column 

```python
# IF the coloum under the do not contact has blank means we will give them a call 
# If Y mean we will not give them a call and we will drop it 
# IF N means we will call them 
```

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
df
```

```python
#Same for phone number 
for x in df.index:
    if df.loc[x,"Phone_Number"] == '':
        df.drop(x, inplace=True)
df
```

#### Resetting the index as for better understanding 

```python
df = df.reset_index(drop = True)
df
```
