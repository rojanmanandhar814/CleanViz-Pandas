### Pandas Visualization 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\BootCamp\Python\Web scraping project pandas\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
df.plot()
```

```python
print(plt.style.available)
plt.style.use('fivethirtyeight')
```

## Plots

```python
df.plot(kind = 'line')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Rating', xlabel = 'Daily Rating', ylabel = 'Scores')
```

### Bar Plots

```python
df.plot(kind = 'bar', stacked = True)
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = True) #as we are only looking at flavor rating data 
```

```python
df.plot.barh(stacked = True)
```

### Scatter Plot

```python
# Need to specify x and y axis 
df.plot.scatter(x = 'Texture Rating', y = 'Flavor Rating', s = 415, c= 'pink') #s is the zooming in of the visualization 
```

### Histogram

```python
df.plot.hist(bins = 45)
```

### Box Plot 

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

## pie chart

```python
df.plot.pie(y = 'Overall Rating', figsize = (10,5))
```
